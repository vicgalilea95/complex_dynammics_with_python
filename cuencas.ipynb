### Funcion de iteracion del metodo de Schröeder aplicado al polinomio $p[z] := (z^2 - 1) (z - \lambda)$

```python
S_z = lambda z, a : (a+z*(4*z**2+z*(-10+z**2)*a+4*a**2))/(1+3*z**4-4*z*(1+z**2)*a+2*(1+z**2)*a**2)
```

```python
import math

z = math.sqrt(3) * 1j
lambda_=math.sqrt(3) * 1j
roots=[1,-1,lambda_]
ratio=10
config={}
config['n_pixels']=3**6
config['x_min']=-ratio
config['x_max']=ratio
config['y_min']=-ratio
config['y_max']=ratio
config['nombre_archivo']='plano_infinito_SCH.png'
config['colors'] = ['magenta','cyan', 'yellow','black']
cmap = mpl.colors.ListedColormap(colors=config['colors'])
roots
```

```python
xreal = np.linspace((config['x_min']), (config['x_max']),int(config['n_pixels']))
ximag = np.linspace(config['y_min'], config['y_max'],int(config['n_pixels']))
Xreal, Ximag = np.meshgrid(xreal, ximag)
X= Xreal+1j*Ximag 
tolerance = 0.0001
max_iter=200
poles=[-0.988781+0.93561j,0.+0.233266j,0.-0.771152j,0.988781+0.93561j]
# Dimensiones de la imagen
width, height = 1280, 720

# Densidad de píxeles
ppp = 73

# Crear la figura con las dimensiones y densidad de píxeles deseadas
fig = plt.figure(figsize=(width / ppp, height / ppp), dpi=ppp)
result_l=list()
for fila in X:
    fila_r=list()
    for elemento in fila:
        r_e=3
        z_0=elemento
        for i in range(max_iter):
            z_0=S_z(z_0,lambda_)
            encontrada=False
            for root in roots:
                if (np.abs(z_0-root)<tolerance):
                    r_e=roots.index(root)
                    encontrada=True
                    break
            if encontrada:
                break   
        fila_r.append(r_e)
    result_l.append(fila_r)
    
cmap = mpl.colors.ListedColormap(config['colors'])
plt.imshow(result_l,extent=(config['x_min'], config['x_max'],config['y_min'], config['y_max']), cmap=cmap)
plt.savefig('landa_sqrt_3_i.jpg',dpi=ppp, bbox_inches='tight')
```

## Familia Chebyshev-Halley con $\infty$ como punto fijo extrano super atractor

#### $\alpha=\frac{5}{2}$, $\infty$ es indiferente

```python
F_z = lambda z : (-2-3*z**2+z**4)/(z*(-5+z**2))
roots=[1,-1]
ratio=4
n_pixels=3**5
tolerance = 0.001
max_iter=100
xreal = np.linspace(-ratio, ratio,n_pixels)
ximag = np.linspace(-ratio, ratio,n_pixels)
Xreal, Ximag = np.meshgrid(xreal, ximag)
X= Xreal+1j*Ximag 
colors = ['magenta','cyan', 'white','black']
width, height = 1280, 720

# Densidad de píxeles
ppp = 73

# Crear la figura con las dimensiones y densidad de píxeles deseadas
fig = plt.figure(figsize=(width / ppp, height / ppp), dpi=ppp)
result_l=list()
for fila in X:
    fila_r=list()
    for elemento in fila:
        r_e=3
        z_0=elemento
        for i in range(max_iter):
            z_0=F_z(z_0)
            encontrada=False
            for root in roots:
                if (np.abs(z_0-root)<tolerance):
                    r_e=roots.index(root)
                    encontrada=True
                    break
                elif (np.abs(z_0))>100:
                    r_e=2
                    encontrada=True
                    break
            if encontrada:
                break   
       # if (r_e==3):
        #    print(elemento,z_0)
        fila_r.append(r_e)
    result_l.append(fila_r)
    
cmap = mpl.colors.ListedColormap(colors)
plt.imshow(result_l,extent=(-ratio, ratio,-ratio, ratio), cmap=cmap)
plt.savefig('foto_infi_indiferente_1.jpg',dpi=ppp, bbox_inches='tight')
    
```
