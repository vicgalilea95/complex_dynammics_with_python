## Búsqueda de ciclos

```python
def cycle_search(list_,tolerance):
    cycle_lenght=0
    aux=list_[-1]
    
    for i in range(2,len(list_)):
        
        if np.abs(aux-list_[-i])<0.001:
            return i-1
        i+=1
    return-1

        
def is_root(polynomial,number):
    if np.abs(polynomial(number))<0.001:
        return True
    return False
```

## Dibujo cuencas de atracción, basins of attraction for Schröder's Method

```python
config={}
config['n_pixels']=3**4
config['x_min']=-0.5
config['x_max']=0.5
config['y_min']=-0.5
config['y_max']=0.5
config['nombre_archivo']='plano_SCH.png'
config['colors'] = ['cyan','magenta','yellow','black']
config['funcion_iteracion']='(-k**3-4*k**2*z-4*k*z**2-6*k**2*z**2-16*k*z**3+12*k**2*z**3-4*z**4+3*k*z**4)/(-2*k**2+3*k**3-8*k*z+12*k**2*z-8*z**2+6*k*z**2-16*z**3+12*k*z**3-12*z**4+9*k*z**4)'

cmap = mpl.colors.ListedColormap(colors=config['colors'])
```

```python
xreal = np.linspace((config['x_min']), (config['x_max']),int(config['n_pixels']))
ximag = np.linspace((config['y_min']), (config['y_max']),int(config['n_pixels']))
Xreal, Ximag = np.meshgrid(xreal, ximag)
X= Xreal+1j*Ximag #1j es la componente imaginaria en Numpy

max_iter=100
f_iter = lambda z : (-1-4*z-10*z**2-4*z**3-z**4)/(-1-4*z+2*z**2+4*z**3+3*z**4)
result_l=list()
for fila in X:
    fila_r=list()
    for elemento in fila:
        
        for i in range(max_iter):
            r_e=3
            elemento=f_iter(elemento)
            if (np.abs((elemento-1j))<0.0001):
                r_e=0
                break
            elif np.abs((elemento+1)) <0.0001:
                r_e=1
            elif np.abs((elemento+1j)) <0.0001:
                r_e=2
            
        
        fila_r.append(r_e)
    result_l.append(fila_r)

cmap = mpl.colors.ListedColormap(config['colors'])
plt.imshow(result_l,extent=(config['x_min'], config['x_max'],config['y_min'], config['y_max']), cmap=cmap)
plt.savefig('preueba_new_1.png', dpi=120)
```
